```python
import numpy as np 
import scipy.io
data = scipy.io.loadmat('iris_multiclass.mat')
x_train_indices=data['indices_train']
x_test_indices=data['indices_test']
FeaturesData=data['meas']
LabelsData=data['species']
X=np.column_stack((FeaturesData,LabelsData))
np.shape(X)
x_train_indices
x_train,x_test=X[x_train_indices-1,:],X[x_test_indices-1,:] #indices from 0 to len-1
np.shape(x_train)
x_train=x_train[0]
x_test=x_test[0]
print(np.shape(x_train))
print(np.shape(x_test))
```

```python
import matplotlib.pyplot as plt
t=np.arange(len(x_train))
plt.figure(6)
plt.subplot(2,2,1)
plt.scatter(t,x_train[:,0],s=100,c='g')

plt.subplot(2,2,2)
plt.scatter(t,x_train[:,1],s=100,c='r')

plt.subplot(2,2,3)
plt.scatter(t,x_train[:,2],s=100,c='y')

plt.subplot(2,2,4)
plt.scatter(t,x_train[:,3],s=100,c='b')

plt.show()
```

```python

for i in range(len(x_train[:,4])):
    if x_train[i,4]=='setosa':
        x_train[i,4]=1
    else:
        x_train[i,4]=2
```

```python
Y=x_train[:,4]
print('STRTING LINEAR REGRESSION')
print('Our label 1 for setosa and 2 for others= ',Y,np.shape(Y))
X=np.column_stack((x_train[:,:4],np.ones(len(x_train))))
```

```python
from numpy.linalg import inv
X=np.column_stack((x_train[:,:4],np.ones(len(x_train))))
X=np.matrix(X, dtype='float')
theta=inv(X.T.dot(X)).dot(X.T).dot(Y)
print(theta)
predict=np.dot(theta,X.T)
t=np.arange(len(X[:,0]))
plt.figure(5)
plt.plot(predict.T)
plt.show()
```

```python
import numpy as np
def lossL2(Y,predict): 
    dis=[]
    for i in range(len(Y)):
        dis.append((Y[i]-predict[i])**2) #numpy np is just for array
    print('Mean Squar Error is = ',np.mean(dis))
```

```python
plt.figure(1)
plt.ylabel('Our two classes,class 1.0 for Setosa,class 2.0 for others')
plt.title('Linear Regression')
plt.plot(t,predict.T,'bo',t,Y,'ro')
plt.show()
lossL2(Y,predict.T)
```

```python
X=np.column_stack((x_train[:,:4]**2,x_train[:,:4],np.ones(len(x_train))))
X=np.matrix(X, dtype='float')
theta=inv(X.T.dot(X)).dot(X.T).dot(Y)
predict=np.dot(theta,X.T)
t=np.arange(len(X[:,0]))
print(len(X[:,0]))
print(np.shape(predict))
plt.plot(predict.T)
plt.show()
plt.figure(2)
plt.ylabel('Our two classes,class 1.0 for Setosa,class 2.0 for others')
plt.title('Plynomial least squar fit with degree 2')
plt.plot(t,predict.T,'bo',t,Y,'ro')
plt.show()
lossL2(Y,predict.T)
```

```python
X=np.column_stack((x_train[:,:4]**3,x_train[:,:4]**2,x_train[:,:4],np.ones(len(x_train))))
X=np.matrix(X, dtype='float')
theta=inv(X.T.dot(X)).dot(X.T).dot(Y)
predict=np.dot(theta,X.T)
t=np.arange(len(X[:,0]))
print(len(X[:,0]))
print(np.shape(predict))
plt.plot(predict.T)
plt.show()
plt.figure(2)
plt.ylabel('Our two classes,class 1.0 for Setosa,class 2.0 for others')
plt.title('Plynomial least squar fit with degree 3')
plt.plot(t,predict.T,'bo',t,Y,'ro')
plt.show()
lossL2(Y,predict.T)
```

```python
X=np.column_stack((x_train[:,:4]**4,x_train[:,:4]**3,x_train[:,:4]**2,x_train[:,:4],np.ones(len(x_train))))
X=np.matrix(X, dtype='float')
theta=inv(X.T.dot(X)).dot(X.T).dot(Y)
predict=np.dot(theta,X.T)
t=np.arange(len(X[:,0]))
print(len(X[:,0]))
print(np.shape(predict))
plt.plot(predict.T)
plt.show()
plt.figure(2)
plt.ylabel('Our two classes,class 1.0 for Setosa,class 2.0 for others')
plt.title('Plynomial least squar fit with degree 4')
plt.plot(t,predict.T,'bo',t,Y,'ro')
plt.show()
lossL2(Y,predict.T)
```

```python
X=np.column_stack((x_train[:,:4]**5,x_train[:,:4]**4,x_train[:,:4]**3,x_train[:,:4]**2,x_train[:,:4],np.ones(len(x_train))))
X=np.matrix(X, dtype='float')
theta=inv(X.T.dot(X)).dot(X.T).dot(Y)
predict=np.dot(theta,X.T)
t=np.arange(len(X[:,0]))
print(len(X[:,0]))
print(np.shape(predict))
plt.plot(predict.T)
plt.show()
plt.figure(2)
plt.ylabel('Our two classes,class 1.0 for Setosa,class 2.0 for others')
plt.title('Plynomial least squar fit with degree 5')
plt.plot(t,predict.T,'bo',t,Y,'ro')
plt.show()
lossL2(Y,predict.T)
```
