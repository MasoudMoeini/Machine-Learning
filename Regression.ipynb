```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(1)
plt.plot(data[:,0])
plt.show()
```

```python
plt.figure(2)
plt.plot(data[:,1])
plt.show()
```

```python
plt.figure(7)
plt.scatter(data[:,0],data[:,1],s=100,c='g')
```

```python
#polynomial least squar fit
#coefficients = numpy.polyfit(x_data, y_data, degree)
#fitted_data = numpy.polyval(coefficients, x_data)
degree=1
coeff = np.polyfit(data[:,0],data[:,1], degree)
Y_fitted = np.polyval(coeff, data[:,0])
plt.figure(8)
plt.plot(data[:,0],Y_fitted, label='linear regression')
plt.plot(data[:,0],data[:,1], label='original data')
plt.grid(True)
plt.suptitle('Linear POLYNOMIAL least squar fit ')
plt.xlabel('input data')
plt.ylabel('results')
plt.legend()
plt.show()
```

```python
degree=2
coeff = np.polyfit(data[:,0],data[:,1], degree)
Y_fitted = np.polyval(coeff, data[:,0])
plt.figure(8)
plt.plot(data[:,0],Y_fitted, label='Prediction with Polynomial degree 2')
plt.plot(data[:,0],data[:,1], label='original data')
plt.grid(True)
plt.suptitle('POLYNOMIAL least squar fit ')
plt.xlabel('input data')
plt.ylabel('results')
plt.legend()
plt.show()
```

```python
#calculating confidence interval
m=np.mean(data[:,1])
sigma2=np.mean((data[:,1]-m)**2)
sigma=np.sqrt(sigma2)
degree=15
coeff = np.polyfit(data[:,0],data[:,1], degree)
Y_fitted = np.polyval(coeff, data[:,0])

plt.figure(figsize=(20, 10))
plt.plot(Y_fitted-sigma)
plt.plot(Y_fitted+sigma)
plt.plot(data[:,1])
plt.plot(Y_fitted, color='y')
plt.show()
```
