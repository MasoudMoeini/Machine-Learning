```python
#Assignment 06.2 (SVM cancer detection, 2+3+1 points)
import scipy.io
data=scipy.io.loadmat('cancer-data.mat')
TrainInput=data['cancerInput_train']
TrainTarget=data['cancerTarget_train']
y=TrainTarget[:500]
a=np.array([])
a=np.append(a,y)
y=a
```

```python
TrainInput.shape
```

```python

np.shape(TrainInput)
plt.scatter(TrainInput[:,0],TrainInput[:,1],c=y,s=50, cmap='autumn')
plt.show()
```

```python
def lossL2(Y,predict):
      return (Y-predict)**2
def lossL3(Y,predict):
    dis=[]
    for i in range(len(Y)):
        dis.append((Y[i]-predict[i])**2) #numpy np is just for array
    return (np.mean(dis))
def truepredict(Y,predict):
    count=0
    for i in range(len(Y)):
        if(Y[i]==predict[i]):
            count+=1
    truepredict=np.dot((count/len(Y)),100)
    print(truepredict,"% of data, predicted successfully")
```

```python
print("prediction when c=0.01")
x=TrainInput[:,:2]
C=[0.01, 0.1, 0.5, 1, 5, 10, 50]
clf = SVC(C=0.01, kernel='linear')
clf.fit(x, y)
pred=np.array([])
predict=[]
for i in range(len(x)): 
    pred=clf.predict([[x[i,0], x[i,1]]])
    predict.append(pred)
    #print("prediction is = ",pred, "error is =", lossL2(y[i],pred))
a=np.array([])
a=np.append(a,predict)
predict=a
t=np.arange(len(y))

#*************************************************************************
plt.figure(1)
plt.ylabel('Our two classes,class -1.0 for Malignant,class 1.0 for benign')
plt.title('Linear Regression with SVM,real data red,predict blue')
plt.plot(t,y,'ro',t,predict,'bo')
print('MSE is',lossL3(y,predict))
truepredict(y,predict)
plt.figure(22)
plt.subplot(1,2,1)
plt.title('real data')
plt.scatter(x[:,0],x[:,1],c=y,s=50, cmap='autumn')
plt.subplot(1,2,2)
plt.title('predicted')
plt.scatter(x[:,0],x[:,1],c=predict,s=50, cmap='autumn')
plt.show()
```

Assignment 06.3 (Computational complexity, 1+1+1 points)

```python
import numpy as np
from numpy.linalg import inv
from numpy.linalg import norm
import scipy.io
data=scipy.io.loadmat('cancer-data.mat')
TrainInput=data['cancerInput_test']
TrainTarget=data['cancerTarget_test']
y=TrainTarget[:500]
a=np.array([])
a=np.append(a,y)
Y=a
X=TrainInput
t=np.arange(len(Y))
X1=np.column_stack((X**8,X**7,X**6,X**5,X**4,X**3,X**2,X,np.ones(len(X))))
theta = inv(X1.T.dot(X1)).dot(X1.T).dot(Y)
predict=np.dot(X1,theta)
plt.figure(29)
plt.subplot(1,2,1)
plt.title('real data')
plt.scatter(X[:,0],X[:,1],c=Y,s=50, cmap='autumn')
plt.subplot(1,2,2)
plt.title('predicted')
plt.scatter(X[:,0],X[:,1],c=predict,s=50, cmap='autumn')
plt.show()
#euclidian distance
dis=np.mean(np.sqrt((Y-predict)**2))
print('MSE is',lossL3(y,predict))
print("Euclidian Distance Error is =",dis)
truepredict(Y,predict)
print('Error is low but Not Perfect 100% percent matched results by using Train Data')
```

```python
#Applying KNN
import numpy as np
from numpy.linalg import inv
from numpy.linalg import norm
import scipy.io
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import cross_val_score
knn=KNeighborsClassifier()
data=scipy.io.loadmat('cancer-data.mat')
TrainInput=data['cancerInput_test']
TrainTarget=data['cancerTarget_test']
y=TrainTarget[:500]
a=np.array([])
a=np.append(a,y)
Y=a
X=TrainInput
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(X,Y)

# for display purposes, we fit the model on the first two components i.e. PC1, and PC2
knn.fit(X[:,0:2],Y)
# Plotting the decision boundary for all data (both train and test)
# Create color maps
from matplotlib.colors import ListedColormap
cmap_light = ListedColormap(['#AAFFAA','#FFAAAA'])
cmap_bold = ListedColormap(['#0000FF','#FF0000'])
# creating a meshgrid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h=0.05
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
xy_mesh=np.c_[xx.ravel(), yy.ravel()]
Z = knn.predict(xy_mesh)
%matplotlib inline
Z = Z.reshape(xx.shape)
#print(Z)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
ax=plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=cmap_bold)
plt.xlim(xx.min(), xx.max());plt.ylim(yy.min(), yy.max())
plt.xlabel('PC1');plt.ylabel('PC2')
plt.title('KNN')
plt.show()
```

```python
from sklearn.grid_search import GridSearchCV
k_range=list(range(1,50))
param_grid=dict(n_neighbors=k_range)
scores = ['accuracy', 'recall'] # evaluation scores for selecting best parameter
for sc in scores:
    grid_knn=GridSearchCV(knn,param_grid,cv=10,scoring=sc,n_jobs=-1)
    print("# Tuning hyper-parameters for %s" % sc)
    grid_knn.fit(X,Y)
    print(grid_knn.best_params_)
    print(np.round(grid_knn.best_score_,5))
```

```python
Assuming k is fixed (as both of the linked lectures do), then your algorithmic choices will determine whether your computation takes O(nd+kn) runtime or O(ndk) runtime.

First, let's consider a O(nd+kn) runtime algorithm:

Initialize selectedi=0 for all observations i in the training set
For each training set observation i, compute disti, the distance from the new observation to training set observation i
For j=1 to k: Loop through all training set observations, selecting the index i with the smallest disti value and for which selectedi=0. Select this observation by setting selectedi=1.
Return the k selected indices
Each distance computation requires O(d) runtime, so the second step requires O(nd) runtime. For each iterate in the third step,
we perform O(n) work by looping through the training set observations, so the step overall requires O(nk) work. 
The first and fourth steps only require O(n) work, so we get a O(nd+kn) runtime.
```
